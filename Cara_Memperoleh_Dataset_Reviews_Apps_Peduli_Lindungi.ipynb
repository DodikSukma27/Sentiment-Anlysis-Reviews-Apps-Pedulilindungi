```python
# import module untuk mendapatakn data play store
from google_play_scraper import app

import pandas as pd

import numpy as np
```

```python
#Scraping data reviewer Pengguna Aplikasi Peduli Lindungi

from google_play_scraper import Sort, reviews

result, continuation_token = reviews(
    'com.telkom.tracencare',
    lang='id', # defaults to 'id' agar bahasa yang tersimpan pada penambangan data yaitu bahasa indonesia
    country='id', # defaults to 'id' agar hanya memfilter data review dari negara indonesia
    sort=Sort.MOST_RELEVANT, # defaults to Sort.MOST_RELEVANT untuk data yang cocok atau Sort.NEWEST untuk data yang terbaru reviews
    count=1000, # defaults to 100
    filter_score_with=None # defaults to None(agar mendapatkan semua score) gunakan 1 atau 2 atau 3 atau 4 atau 5 untuk memilih score bintang di playstore
)
```

```python
df_busu = pd.DataFrame(np.array(result),columns=['review'])

df_busu = df_busu.join(pd.DataFrame(df_busu.pop('review').tolist()))

df_busu.head()
```

```python
len(df_busu.index) # menhitung jumlah data yang kita dapat
```

```python
df_busu[['userName', 'score','at', 'content']].head()  #mengambil hanya userName, rating, date-time, dan reviews 
```

```python
new_df = df_busu[['userName', 'score','at', 'content']]
sorted_df = new_df.sort_values(by='at', ascending=False) #Sort by Newst, change to True if you want to sort by Oldest.
sorted_df.head()
```

```python
my_df = sorted_df[['userName', 'score','at', 'content']] #get userName, rating, date-time, and reviews only
```
